## Labels Legend Title Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline

day = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(day,max_t,label = "Max")
plt.plot(day,min_t, label = "Min")
plt.plot(day,avg_t, label = "Avg")

plt.legend(shadow = True,fontsize = "large")
# plt.legend(loc = "best",shadow = True)
# plt.legend(loc = "lower right",shadow = True)
# plt.legend(loc = "upper right",shadow = True)
# plt.legend(loc = 7,shadow = True)

plt.grid()
```

# BAR CHART

```python
import matplotlib.pyplot as plt
import numpy as np

company = ['GOOGL','AMZN','MSFT','META']
revenue = [92,56,72,58]
profit = [47,21,43,12]

plt.ylabel("Billion Dollar")
plt.title("US Tech Company's Net Revenue")

xpos = np.arange(len(company))
plt.xticks(xpos,company)

# plt.plot(company,revenue)
# plt.plot(company,profit)

plt.bar(xpos-0.2,revenue,width = 0.4,label = "Revenue(BLN)")
plt.bar(xpos+0.2,profit,width = 0.4,label = "Profit")

# plt.bar(company,revenue,label = "Revenue")
# plt.bar(company,profit,label = "Profit")

plt.legend()
```

## Pie Chart

```python
import matplotlib.pyplot as plt

exp_vals = [1800, 3000, 500, 240, 1500]
exp_labels = ['Home Rent','Food','Phone bill','Travel Bus','Other']

plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels,
        radius = 1.5, autopct = '%0.1f%%',
        shadow = True,explode = [0.2,0,0,0,0])

plt.show()
```

```python
import matplotlib.pyplot as plt

exp_vals = [1800, 3000, 500, 240, 1500]
exp_labels = ['Home Rent','Food','Phone bill','Travel Bus','Other']

plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels,
        radius = 1.5, autopct = '%0.1f%%',
        shadow = True,explode = [0.2,0,0,0,0],
       startangle= -45)

plt.show()
```

## Divya Bhaskar Paper 17 OCT
