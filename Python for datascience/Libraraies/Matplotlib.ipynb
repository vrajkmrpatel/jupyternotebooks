### subplot()

The Matplotlib subplot() function is defined as to plot two or more plots in one figure. We can use this method to separate two graphs which plotted in the same axis Matplotlib supports all kinds of subplots, including 2x1 vertical, 2x1 horizontal, or a 2x2 grid.

It accepts the three arguments: they are nrows, ncols, and index. It denote the number of rows, number of columns and the index.


The subplot() function can be called in the following way:

subplot(nrows,ncols,index,**kwargs)  
subplot(pos,**kwargs)     
subplot(ax)  

subplot (1, 3, 2) is the same as the subplot (132).


```python
from matplotlib import pyplot as plt
names = ['Abhishek', 'Himanshu', 'Devansh']  
marks= [87,50,98]  
  
plt.figure(figsize=(9,3))  
  
plt.subplot(131)  
plt.bar(names, marks)  
plt.subplot(132)  
plt.scatter(names, marks)  
plt.subplot(133)  
plt.plot(names, marks)  
plt.suptitle('Categorical Plotting')  
plt.show()  
```

## Different types of graph

1) line graph

```python
import matplotlib.pyplot as plt

x = [16,8,10]
y = [8,12,28]
x2 = [8,15,7]
y2 = [5,18,20]

plt.xlabel("x line")
plt.ylabel("y line")
plt.title("Line Graph")

plt.plot(x,y,'r',label ='line one',linewidth = 3)
plt.plot(x2,y2,'g',label = 'line two')
plt.legend()
plt.grid(True,color = 'k')
plt.figure()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

ax = plt.axes()
x = np.linspace(0,10,1000)

ax.plot(x,np.sin(x))
plt.show()
```

```python

import matplotlib.pyplot as plt  
import numpy as np  
  
x = np.arange(0.0, 2, 0.01)  
y1 = np.sin(2 * np.pi * x)  
y2 = 1.2 * np.sin(4 * np.pi * x)  
fig, ax = plt.subplots(1, sharex=True)  
ax.plot(x, y1, x, y2, color='black')  
ax.fill_between(x, y1, y2, where=y2 >= y1, facecolor='blue', interpolate=True)  
ax.fill_between(x, y1, y2, where=y2 <= y1, facecolor='red', interpolate=True)  
ax.set_title('fill between where')
```

2. Bar graph

```python
player = ['Virat','Rohit','Dhoni','Raina']
runs = [87,45,56,21]

plt.xlabel("Player name")
plt.ylabel("Runs")
plt.title("Score")
plt.bar(player,runs,color ='m')
```

```python
player = ['Virat','Rohit','Dhoni','Raina']
runs = [87,45,56,21]

plt.ylabel("Player name")
plt.xlabel("Runs")
plt.title("Score")
plt.barh(player,runs,color ='green')
```
