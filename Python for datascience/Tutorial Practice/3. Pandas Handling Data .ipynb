```python
import pandas as pd
df = pd.read_csv("new_weather.csv",parse_dates =["day"] )
df
```

```python
df.set_index('day',inplace =True)
```

```python
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
#here above method fillna will replace all Nan with o 
#but weather should not be 0 
#so here we will pass dictionary to replace Nan for particular column
```

```python
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'weather': 'no event'
})
new_df
```

```python
new_df = df.fillna(method = "ffill")
new_df
```

```python
new_df = df.fillna(method = "bfill")
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```
